Adding outliers to test the validity of the data set

```python
#Add outliers:
import random
num_outliers = 13
outliers = []

#random values
platforms = ["TikTok", "Facebook", "Telegram", "YouTube"]
locations = ["North Korea", "Japan", "Germany", "Australia"]
genders = ["Female", "Male"]
demographics = ["Rural", "Urban", "Sub_Urban"]
professions = ["Unemployed", "Farmer", "Cashier", "driver"]
video_categories = ["ASMR", "Gaming", "Lifestlye"]
watch_reasons = ["Addiction", "Bad Sleeping", "Habit"]
devices = ["Windows", "Tablet", "Quantum Computer"]
oses = ["Linux", "Android", "Windows 95"]
connections = ["Satellite", "Wi-Fi", "Mobile Data"]
frequency=["Night","Afternoon","Evening"]
watch_time=["9:00 pm","5:00 pm","2:00 am"]
current_activity=["Commuting","At home","At work"]

for _ in range(num_outliers):
    outliers.append({
        "age": random.randint(80, 90),  
        "income": random.randint(22000, 25000),  
        "total time spent": random.uniform(600, 720),  
        "number of sessions": random.randint(18, 20),  
        "video length": random.uniform(60, 200),  
        "importance score": random.uniform(0, 1),  
        "time spent on video": random.uniform(50, 200),  
        "number of videos watched": random.randint(60, 70),  
        "scroll rate": random.uniform(100, 500),  
        "productivityloss": random.uniform(8, 10),  
        "satisfaction": random.uniform(0, 2), 
        "self control": random.uniform(0, 1),
        "frequency": random.choice(frequency),
        "platform": random.choice(platforms),
        "watch time": random.choice(watch_time),
        "currentactivity": random.choice(current_activity),
        "location": random.choice(locations),
        "gender": random.choice(genders),
        "demographics": random.choice(demographics),
        "profession": random.choice(professions),
        "owns property": random.choice([True, False]),
        "indebt": random.choice([True, False]),
        "video category": random.choice(video_categories),
        "watch reason": random.choice(watch_reasons),
        "devicetype": random.choice(devices),
        "os": random.choice(oses),
        "connectiontype": random.choice(connections)
    })
outliers_df = pd.DataFrame(outliers)
merged_df = pd.concat([merged_df, outliers_df], ignore_index=True)

merged_df
```

**Hypothesis**

**Hypotheses (1): Effect of Device Type on Watch Time**

Null Hypothesis (H₀): There is no difference in the average watch time between users on mobile devices and desktop devices.

Alternative Hypothesis (H₁): There is a difference in the average watch time between mobile and desktop user

```python
from scipy.stats import ttest_ind

mobile_watch = merged_df[merged_df['devicetype'] == 'Smartphone']['total time spent']
desktop_watch = merged_df[merged_df['devicetype'] == 'Computer']['total time spent']

t_stat, p_val = ttest_ind(mobile_watch, desktop_watch, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("H0:Refused,As there is a difference between the average, H1:accepted  ")
else:
    print("H0:Accepted,There is no difference between the average")
```

**Hypotheses (2):Do Males Spend More Time Watching Videos Than Females**

Null Hypothesis (H₀): The average time spent on social media is equal between males and females.

Alternative Hypothesis (H₁): The average time spent on social media is different between males and females.



```python
male_time = merged_df[merged_df['gender'] == 'Male']['total time spent']
female_time = merged_df[merged_df['gender'] == 'Female']['total time spent']

t_stat, p_val = ttest_ind(male_time, female_time, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("H0:Refused,As The average time spent on social media is not equal between males and females., H1:accepted  ")
else:
    print("H0:Accepted,The average time spent on social media is equal between males and females. ")
```
