```python
import cv2
import glob
import os
import numpy as np

image_folder = "/content/drive/MyDrive/Colab Notebooks/Human Face Expression.v20i.yolov8/train"
image_paths = glob.glob(os.path.join(image_folder, "*.jpg")) + \
              glob.glob(os.path.join(image_folder, "*.png"))

kernel = np.ones((5, 5), np.float32) / 25

for path in image_paths:

    img = cv2.imread(path)
    if img is None:
        print(f"no correct path:  {path}")
        continue


    filtered = cv2.filter2D(img, -1, kernel)

    cv2.imwrite(path, filtered)
```
